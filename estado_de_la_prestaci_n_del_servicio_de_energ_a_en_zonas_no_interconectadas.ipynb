```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
from unidecode import unidecode


url1 = '/content/drive/MyDrive/Colab Notebooks/proyectos/Estado_de_la_prestación_del_servicio_de_energía_en_Zonas_No_Interconectadas_20260122.csv' #ZNI

url2 ='/content/drive/MyDrive/Colab Notebooks/proyectos/DIVIPOLA_-_Códigos_cabeceras_-_Centros_poblados_20260126.csv' #CODIGOS OFICIALES DE TODAS LAS LOCALIDADES DE COLOMBIA



```

```python
df_ZNI = pd.read_csv(url1,
                     decimal=',',
                     thousands= '.',
                     parse_dates=["FECHA DE DEMANDA MÁXIMA"],
                     date_format= "%Y %b %d %I:%M:%S %p")
df_ZNI
```

```python
df_centros_poblados = pd.read_csv(url2,
                                  decimal=',').drop(2)

df_centros_poblados
```

##**LIMPIEZA DE DATOS**

```python
df_ZNI.rename(columns= {'ID DEPATAMENTO': 'ID DEPARTAMENTO'}, inplace=True) #CORRECCIÓN DE NOMBRE DE LA COLUMNA DEPARTAMENTO
```

```python
df_ZNI.dropna(inplace=True) #ELIMINAR DATOS NULOS
```

```python
df_ZNI['DÍA DE DEMANDA MÁXIMA'] = df_ZNI['DÍA DE DEMANDA MÁXIMA'].apply(lambda x: unidecode(x).upper()) #corrección de tildes y mayúsculas, normalizando el texto.
```

```python
def convertir_float(valor): #función para pasar datos de objects a float

  if isinstance(valor, (int, float)) is True: #primer ciclo comprueba si es int o float, en el caso de que sí, devuelve el valor
    return float(valor)

  elif ( '.'  not in valor ) and ( ',' not in valor ): ##verifica que si no tiene coma ni punto, solo lo convierte en float
    return float(valor)

  elif ( '.' in valor ) and ( ',' in valor ): #verifica si tiene punto y coma a la vez, donde primero elmina los puntos y luego las comas las remplaza por ppuntos
    valor = valor.replace('.', '')
    valor = valor.replace(',', '.')
    return float(valor)

  elif  ',' in valor:  # si solo tiene una coma, la transforma en punto.
    valor = valor.replace(',', '.')
    return float(valor)

  else:
    return np.nan
```

```python
#se renombran algunas de las columnas de los centros poblados.
df_centros_poblados.rename(columns= {'Código Departamento' : 'ID DEPARTAMENTO',
                            'Codigo Municipio' : 'ID MUNICIPIO',
                            'Código Centro Poblado' : 'ID LOCALIDAD',
                            'Tipo*' : 'TIPO',
                            'longitud' : 'LONGITUD',
                            'Latitud' : 'LATITUD'},
                          inplace=True)
```

```python
#Se unen ambos dataframes
df_ZNI_limpio = pd.merge(df_ZNI,
                          df_centros_poblados,
                          how= 'left',
                          on= ['ID DEPARTAMENTO', 'ID MUNICIPIO', 'ID LOCALIDAD']).dropna() #se eliminan los valores nulos ya que estos no coinciden con las localidades oficiales registradas.
```

```python
df_ZNI_limpio.drop(columns = ['DEPARTAMENTO', 'MUNICIPIO', 'LOCALIDAD'], inplace=True)

#Extraer filas para ser reubicadas
DEPARTAMENTOS = df_ZNI_limpio.pop("Nombre_departamento")
MUNICIPIOS = df_ZNI_limpio.pop("Nombre Municipio")
LOCALIDADES = df_ZNI_limpio.pop("Nombre Centro Poblado")

#reinsertar las filas en sus posiciones originales
df_ZNI_limpio.insert(1, 'DEPARTAMENTO', DEPARTAMENTOS)
df_ZNI_limpio.insert(3, 'MUNICIPIO', MUNICIPIOS)
df_ZNI_limpio.insert(5, 'LOCALIDAD', LOCALIDADES)
```

```python
#Recorre todas las columnas que hay que limpiar en el dataframe, pasandolas todas a floats
columnas_para_limpiar = ['ENERGÍA ACTIVA', 'ENERGÍA REACTIVA', 'POTENCIA MÁXIMA', 'PROMEDIO DIARIO EN HORAS', 'LONGITUD', 'LATITUD']
for col in columnas_para_limpiar:
    df_ZNI_limpio[col] = df_ZNI_limpio[col].apply(convertir_float)
```
